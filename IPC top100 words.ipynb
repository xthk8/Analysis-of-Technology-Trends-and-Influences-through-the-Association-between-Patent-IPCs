맨위에 코드 단락 한번 누르면 top_100-words_all_ipc csv 파일이 생겨 그걸 보고 하면돼

```python
import pandas as pd

# 1. 각 ipc별로 tf_idf 상위 100개 단어 추출하는 함수
def get_top_100_words(df):
    top_100_words_by_ipc = pd.DataFrame()
    ipc_codes = df['ipc'].unique()
    
    for ipc_code in ipc_codes:
        ipc_df = df[df['ipc'] == ipc_code]
        top_100_words = ipc_df.sort_values(by='tf_idf', ascending=False).head(100)
        top_100_words['ipc'] = ipc_code  # IPC 코드를 추가
        top_100_words_by_ipc = pd.concat([top_100_words_by_ipc, top_100_words], ignore_index=True)

    return top_100_words_by_ipc

# 2. 단어 삭제 및 기록 함수
def delete_words_and_log(df, words_to_delete, log_filename="deleted_words_log.csv"):
    deleted_log = pd.DataFrame(columns=['ipc', 'word'])

    # 삭제할 단어들에 대해 삭제 기록
    for word in words_to_delete:
        deleted_entries = df[df['word'] == word][['ipc', 'word']]  # 삭제될 단어의 ipc 및 단어 정보 추출
        deleted_log = pd.concat([deleted_log, deleted_entries], ignore_index=True)
    
    # 로그 파일에 추가 기록
    deleted_log.to_csv(log_filename, mode='a', header=False, index=False, encoding='utf-8')
    
    # 원본 데이터에서 단어 삭제
    updated_df = df[~df['word'].isin(words_to_delete)]
    return updated_df

# 1. 원본 데이터 불러오기
df = pd.read_csv("C:/Users/한고은/Desktop/ipc/combined_split_ipc_words.csv", encoding='utf-8')

# 2. 상위 100개 단어 추출 (IPC별로 한꺼번에 처리)
top_100_words_by_ipc = get_top_100_words(df)

# 상위 100개 단어를 CSV로 저장 (cp949 인코딩)
top_100_words_filename = "top_100_words_all_ipc.csv"
top_100_words_by_ipc.to_csv(top_100_words_filename, index=False, encoding='utf-8')
print(f"Top 100 words for all IPCs have been saved to {top_100_words_filename}")

```

이제 여기 아래 단락에다가 지울 단어를 입력하면 원래 원본데이터랑 top_100_words_all_ipc 의 단어가 업데이트 되거든?

그러면 여기서 계속 단어 바꿔가면서 돌려주기만 하면돼

```python

# 3. 작업자가 삭제할 단어 리스트를 직접 정의
# 예시로 삭제할 단어를 미리 지정
words_to_delete = ["휘발시키기"]

# 4. 삭제할 단어를 원본 데이터 프레임에서 삭제하고 기록 (deleted_words_log.csv에 기록)
df = delete_words_and_log(df, words_to_delete)

# 삭제 후 다시 저장
df.to_csv("combined_split_ipc_words.csv", index=False, encoding='utf-8')

# 5. 삭제 후 다시 상위 100개 단어 추출 (전체)
updated_top_100_words_by_ipc = get_top_100_words(df)
print(f"\nUpdated Top 100 words after deletion:")
print(updated_top_100_words_by_ipc[['ipc', 'word', 'tf_idf']].to_string(index=False))

# 삭제 후 다시 상위 100개 단어 저장
updated_top_100_words_filename = "top_100_words_all_ipc.csv"
updated_top_100_words_by_ipc.to_csv(updated_top_100_words_filename, index=False, encoding='utf-8')
print(f"\nUpdated top 100 words after deletion have been saved to {updated_top_100_words_filename}")

print("모든 작업이 완료되었습니다. 원본 CSV 파일과 로그 파일이 업데이트되었습니다.")
```
